### Display countires that share border with India

```python
def isIndianBorder(lc):
    for c in lc:
         if c == "IND":
            return True
    
    return False
```

```python
def hasIndiaAsBorder(lc):
    return 'IND' in lc
```

```python
df.loc [ df['borders'].apply(isIndianBorder) , ['name','borders']]
```

```python
df.loc [ df['borders'].apply(lambda l: 'IND' in l) , ['name','borders']]
```
