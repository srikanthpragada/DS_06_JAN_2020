### Broadcasting

```python
a = np.array([1,2,3,4,5])
```

```python
a + 5    # Value 5 is broadcast to an array of 5 elements because a is an array of 5 elements 
```

```python
a1 = np.array ([1,2,3])
a2 = np.array ([[1,2,3],[4,5,6]])
a1 + a2
```

```python
# Error as no. of cols is not same 
a1 = np.array ([1,2])   # 2 cols
a2 = np.array ([[1,2,3],[4,5,6]])   # 3 cols 
a1 + a2
```

```python
# No. of cols is not same so broadcast array of 1 col to 3 cols 
a1 = np.array ([1,2]).reshape(2,1)   # 2,1 dim
a2 = np.array ([[1,2,3],[4,5,6]])   # 2,3 cols 
a1 + a2
```
